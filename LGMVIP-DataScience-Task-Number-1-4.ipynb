## Importing Libraries

```python
import numpy as np
import pandas as pd
import PIL
from PIL import Image, ImageChops, ImageFilter
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import cv2
from IPython.display import display, Image as ipimage
```

## Using PIL library:

```python
# Opening color image using Image.open
color_image=Image.open('/kaggle/input/letsgrowmoreimage/istockphoto-1309405105-612x612.jpg')
color_image
```

```python
# Converting Color image to Gray Image using 'convert('L')'
gray_image=color_image.convert('L')
gray_image
```

```python
# Inverting Image means substracting individual pixel values from 255,
# Here it is done using ImageChops.invert() function from PIL library
inverted_image1=ImageChops.invert(gray_image)
inverted_image1
```

```python
# Using lambda and eval function, every pixel value is substracted from 255 to get inverted image
inverted_image=Image.eval(gray_image,lambda x:255-x)
inverted_image
```

```python
np.array(inverted_image)
```

```python
# Applying Blurring to image using 'GaussianBlur' ImageFilter
inverted_image_blurred=inverted_image.filter(ImageFilter.GaussianBlur(radius=5))
```

```python
inverted_image_blurred
```

```python
np.array(inverted_image_blurred)
```

```python
#dividing Gray Image Pixel Values by Inverte Image Pixel Values
pencil_sketch=np.array(gray_image)/np.array(inverted_image_blurred)
```

```python
pencil_sketch
```

```python
#Converting floating pixel values to 'uint8' data type so as to be compatible for plotting image
pencil_sketch_uint=pencil_sketch.astype('uint8')
```

```python
pencil_sketch_uint
```

```python
# Scaling Image by multiplying individual pixel values with 255 to get pencil sketch image
Image.fromarray(255*pencil_sketch_uint)
```

## Using CV2 Library

```python
color_imagecv=cv2.imread('/kaggle/input/letsgrowmoreimage/istockphoto-1309405105-612x612.jpg')
color_imagecv
```

```python
gray_imagecv=cv2.cvtColor(color_imagecv,cv2.COLOR_BGR2GRAY)
gray_imagecv
```

```python
inverted_imagecv=255-gray_imagecv
inverted_imagecv
```

```python
pencil_sketchcv=gray_imagecv/inverted_imagecv
pencil_sketchcv
```

```python
pencil_sketchcv_uint=pencil_sketchcv.astype('uint8')
```

```python
pencil_sketchcv_uint
```

```python
plt.imshow(pencil_sketchcv_uint*255, cmap='gray')
```
